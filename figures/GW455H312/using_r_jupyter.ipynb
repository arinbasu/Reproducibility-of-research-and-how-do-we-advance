## How to use R in Jupyter notebook

In order to use R in jupyter notebook, you will first need to intall an Irkernel in the R terminal application and then start a new jupyter instance. You can also interactively change your R and python instances in jupyter. 

### How to use Jupyter notebook for blogging with hugo, git, and netlify

1. Install node and using npm, install hugo from https://gohugo.io/getting-started/installing/ (see instructions here)
2. Install git and create accounts in both github and netlify
3. Create an empty git repo for the blog
4. Do the writing and analysis in jupyter notebook
5. push the ipynb file using nbconvert to markdown first 
6. Now append the markdown file to a hugo new post

So do something like 

```
hugo new posts/first-post.md
```
Then add the front matter to this new post, such as

```---
title = "My new post" & other meta data
---
```

In the bash shell, do something like

```
cat jupyter_made.md > first-post.md
```

Then generate the hugo site, or build hugo site

Push to github

In Netlify, set up the github to netlify pulling and building of sites and you can then continue to work on the jupyter keeping this clean and pushing the elements you need to hugo and build your static site. 

The idea is to keep the hugo site generated and updated and pushed to github separate from the notebook written and codes written here.
